```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

```python
dataset = pd.read_csv('income.data.csv')
```

```python
print(dataset.shape)
print(dataset.head())
print(dataset.describe())
```

```python
x = dataset.income.values
y = dataset.happiness.values
X_train, X_test, y_train, y_test = train_test_split(x, y,test_size=0.20,random_state=33)
plt.scatter(x, y)
plt.title('Income vs Happiness')
plt.xlabel('Income')
plt.ylabel('Happiness Score')
plt.show()
```

```python
reg = LinearRegression()
reg.fit(np.expand_dims(X_train, axis=-1), y_train)
print('slope is-', reg.coef_[0])
print('Intercept is-', reg.intercept_)
```

```python
plt.scatter(dataset['income'], dataset['happiness'])
x = np.linspace(0,8,100)
y = reg.coef_*x+reg.intercept_
plt.plot(x, y, '-r')
plt.title('Income vs Happiness')
plt.xlabel('Income')
plt.ylabel('Happiness Score')
plt.show()
```

```python
y_pred = reg.predict(np.expand_dims(X_test, axis=-1))
MSE = np.sum((y_test - y_pred)**2)/len(y_test)
print(f'Mean Squred Error :: {MSE}')
plt.scatter(X_test, y_test, color="blue")
plt.plot(X_test, y_pred, color="red", linewidth=3)
```
