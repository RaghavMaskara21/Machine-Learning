```python
import numpy as np
import random,os
```

```python
def ptron(inp1,inp2,outp):
    outp_pn = inp1*weights[0]+inp2*weights[1]+bias*weights[2]
    outp_pn = 1.0/(1+np.exp(-outp_pn))   
    err = outp - outp_pn
    weights[0] += err*inp1*lr   
    weights[1] += err*inp2*lr
    weights[2] += err*bias*lr
```

```python
lr = 1
bias = 1
weights = list()
for k in range(3):
    weights.append(random.random())  

for i in range(50):   
    ptron(0,0,0)     
    ptron(1,1,1)
    ptron(1,0,1)
    ptron(0,1,1)
    print(bias, weights)
```

```python
for x,y in [(0,0),(1,0),(0,1),(1,1)]:
    outp_pn = x*weights[0]+y*weights[1]+bias*weights[2]
    outp = 1.0/(1+np.exp(-outp_pn))
    if outp > 0.5:
      outp = 1
    else:
      outp = 0
    print(x,"OR",y,"yields:",outp)
```
