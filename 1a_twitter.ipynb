```python
# -*- coding: utf-8 -*-
"""Twitter.ipynb

Automatically generated by Colaboratory.

Original file is located at
    https://colab.research.google.com/drive/1Ch67tZyhUwBtZdqXrk0gwVNml3Td1hFC
"""

import tweepy
import pandas as pd
import time

consumer_key = ""
consumer_secret = ""
access_token = ""
access_token_secret = ""
auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)
api = tweepy.API(auth,wait_on_rate_limit=True)

text_query = 'Crypto'
count = 1000
try:

 tweets = tweepy.Cursor(api.search,q=text_query).items(count) 
 
 tweets_list = [[tweet.created_at, tweet.id, tweet.text, tweet.retweet_count, tweet.favorite_count, tweet.user.followers_count, tweet.user.friends_count, tweet.user.verified, tweet.user.statuses_count, tweet.user.listed_count ] for tweet in tweets] 

 tweets_df = pd.DataFrame(tweets_list)
 
except BaseException as e:
    print('failed on_status,',str(e))
    time.sleep(3)

tweets_df.rename(columns={0: 'created_at', 1: 'tweet_id', 2:'text', 3: 'retweet_count', 4:'favorite_count', 5:'user_followers_count', 6:'user_friends_count' , 7:'user_verified' , 8:'user_statuses_count', 9:'user_listed_count'}, inplace=True)

tweets_df.head()
len(tweets_df)
tweets_df.to_csv('crypto_tweets.csv')
```
